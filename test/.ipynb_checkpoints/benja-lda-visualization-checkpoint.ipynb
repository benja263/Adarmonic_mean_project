```python
from helpers import *
import numpy as np
import pandas as pd
import os
import pickle
from LDA_helpers import process_tweets
import re
from pre_process_text import clean
import gensim
from gensim import corpora, models
import itertools
import pyLDAvis
import pyLDAvis.gensim
import matplotlib.pyplot as plt
```

```python
DATA_PATH = access_folder()
GENERATED_DATA_PATH = access_folder('generated')
```

```python
# loading all data
df_s = []
for file in os.listdir(DATA_PATH):
    if file[-4:] == '.csv' and file[:3] == 'IRA':
        df_s.append(pd.read_csv(DATA_PATH + file))
tweets = pd.concat(df_s, sort = False , axis = 0)
tweets = tweets[tweets['language'] == 'English']
tweets.head()
```

```python
categories = tweets.groupby('account_category')
popular_authors = dict()
for name, group in categories:
    popular_authors[name] = group.groupby('author').agg({"followers": "max"}).sort_values('followers', ascending = False).head(4)
```

```python
popular_authors
```

```python
# load model
MODEL_DATA_PATH = access_folder('models')
lda_model = gensim.models.LdaMulticore.load(MODEL_DATA_PATH + 'full_model')
dictionary = corpora.Dictionary.load(MODEL_DATA_PATH + 'full_model.id2word')
```

```python
# take author and find all its tweets
def author_tweets(data, author, category=None):
    if category:
        tweets = data[(data['author'] == author) & (data['account_category'] == category)].content
    else:
        tweets = data[data['author'] == author].content
    return tweets

def clean_tweets(tweets):
    cleaned_tweets = []
    for tweet in tweets:
        cleaned_tweets.append(clean(tweet).split(" "))
    return cleaned_tweets

def topic_prob_to_topics(topics, nb_topics=1):
    best_topic = []
    for topic in topics:
        if topic:
            best, prob = zip(*topic)
            best_topic.append(int(best[0]))
        else:
            best_topic.append(np.nan)
    return best_topic

# convert tweets to corpus
def plot_topic_id_per_author(data, category ,dictionary, model, min_prob=0.5):
    corpus = [dictionary.doc2bow(text) for text in data[category]['Tweets']]
    topics = model.get_document_topics(corpus, minimum_probability=0.4)
    topic_ids = topic_prob_to_topics(topics)
    data[category]['topic_id'] = topic_ids
    data[category].dropna(axis=0, inplace=True)
    groups = data[category][['Author', 'topic_id']].groupby('Author')
    f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2,sharex=True,figsize = [8,8])
    axes = ((ax1, ax2), (ax3, ax4))
    i = 0
    j = 0
    for name, group in groups:
        axes[i][j].set_title(name, fontsize=8)
        axes[i][j].set_xlabel('Topic Id')
        axes[i][j].set_xticks([0, 5,10, 15,20], minor=True)
        axes[i][j].minorticks_on()
        group.plot.hist(bins=30, ax=axes[i][j], legend=False)
        if i == 0 and j == 0:
            j = 1
        elif i == 0 and j == 1:
            i = 1
            j = 0
        elif i == 1 and j == 0:
            j = 1
    f.suptitle(category, fontsize=12)
    f.tight_layout()
```

```python
popular_cleaned_tweets_of_popular_authors = dict()
for category in popular_authors.keys():
    authors = []
    cleaned_tweets = []
    for author in popular_authors[category].index.values:
        author_extracted_tweets = author_tweets(tweets, author, category)
        cleaned = clean_tweets(author_extracted_tweets)
        for tweet in cleaned:
            if tweet[0] != "":
                authors.append(author)
                cleaned_tweets.append(tweet)
    popular_df = pd.DataFrame()
    popular_df['Author'] = authors
    popular_df['Tweets'] = cleaned_tweets
    popular_cleaned_tweets_of_popular_authors[category] = popular_df
```

```python
plot_topic_id_per_author(popular_cleaned_tweets_of_popular_authors, 'Fearmonger' ,dictionary, lda_model, min_prob=0.5)
```

```python
plot_topic_id_per_author(popular_cleaned_tweets_of_popular_authors, 'Commercial' ,dictionary, lda_model, min_prob=0.5)
#LeftTroll, RightTroll, NewsFeed, HashTagGamer, Unknown
```
